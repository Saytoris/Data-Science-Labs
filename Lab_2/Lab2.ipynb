```python
# 1. Імпортуємо бібліотеку
import numpy as np

# --- CREATE (Створення) ---
print("--- CREATE ---")
# Cкаляр (одне значення)
scalar = np.array(10)
print(f"Скаляр:\n{scalar}\n")

# Вектор (одновимірний масив) з arange
vector_arange = np.arange(5, 15, 2) # Від 5 до 15 (не включно) з кроком 2
print(f"Вектор (arange):\n{vector_arange}\n")

# Матриця (двовимірний масив) з random
matrix_random = np.random.randint(0, 10, size=(3, 4)) # Матриця 3x4 з випадковими числами від 0 до 9
print(f"Матриця (random):\n{matrix_random}\n")


# --- READ (Читання) ---
print("--- READ ---")
# Читання скаляра
print(f"Читання скаляра: {scalar.item()}") # .item() для отримання значення Python

# Читання елемента вектора (наприклад, 2-й елемент)
print(f"2-й елемент вектора: {vector_arange[1]}") # Індексація з 0

# Читання елемента матриці (наприклад, 1-й рядок, 3-й стовпець)
print(f"Елемент [1, 2] матриці: {matrix_random[1, 2]}")

# Читання цілого рядка (наприклад, 2-й рядок)
print(f"2-й рядок матриці: {matrix_random[1, :]}")


# --- UPDATE (Оновлення) ---
print("--- UPDATE ---")
# Оновлення вектора (змінимо 2-й елемент на 99)
vector_arange[1] = 99
print(f"Оновлений вектор:\n{vector_arange}\n")

# Оновлення матриці (змінимо елемент [1, 2] на 77)
matrix_random[1, 2] = 77
print(f"Оновлена матриця:\n{matrix_random}\n")


# --- DELETE (Видалення) ---
# Примітка: np.delete створює новий масив, а не змінює старий
print("--- DELETE ---")
# Видалення елемента з вектора (наприклад, 2-й елемент)
vector_deleted = np.delete(vector_arange, 1)
print(f"Вектор з видаленим елементом:\n{vector_deleted}\n")

# Видалення рядка з матриці (наприклад, 1-й рядок, axis=0)
matrix_row_deleted = np.delete(matrix_random, 0, axis=0)
print(f"Матриця з видаленим 1-м рядком:\n{matrix_row_deleted}\n")

# Видалення стовпця з матриці (наприклад, 2-й стовпець, axis=1)
matrix_col_deleted = np.delete(matrix_random, 1, axis=1)
print(f"Матриця з видаленим 2-м стовпцем:\n{matrix_col_deleted}\n")
```

```python
# 1. Імпортуємо бібліотеку
import pandas as pd

# --- Створення DataFrame ---
print("--- Створення DataFrame ---")
# Спосіб 1: Зі словника Python 
data_dict = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emily'],
    'Age': [25, 30, 35, 28, 22],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']
}
df_from_dict = pd.DataFrame(data_dict)
print(f"DataFrame зі словника:\n{df_from_dict}\n")

# Спосіб 2: З масиву NumPy
data_np = np.array([
    [1, 'Product A', 100],
    [2, 'Product B', 150],
    [3, 'Product C', 200]
])
df_from_np = pd.DataFrame(data_np, columns=['ID', 'Product', 'Price'])
print(f"DataFrame з NumPy:\n{df_from_np}\n")


# --- Маніпуляції з даними ---
print("--- Маніпуляції з даними (на базі df_from_dict) ---")

# .head() - показати перші N рядків (за замовчуванням 5) [cite: 61]
print(f"Метод .head(3):\n{df_from_dict.head(3)}\n")

# .describe() - отримати статистичний опис (тільки для числових стовпців) [cite: 64]
print(f"Метод .describe():\n{df_from_dict.describe()}\n")

# .iloc[] - вибірка за числовим індексом 
# (наприклад, 2-й рядок (індекс 1) та всі стовпці)
print(f"Метод .iloc[1]:\n{df_from_dict.iloc[1]}\n")
# (наприклад, перші 3 рядки та стовпці з 0 по 1)
print(f"Метод .iloc[0:3, 0:2]:\n{df_from_dict.iloc[0:3, 0:2]}\n")

# .loc[] - вибірка за міткою (індексом) та назвами стовпців [cite: 69]
# (наприклад, рядок з індексом 0, стовпці 'Name' та 'City')
print(f"Метод .loc[0, ['Name', 'City']]:\n{df_from_dict.loc[0, ['Name', 'City']]}\n")
# (наприклад, фільтрація: всі рядки, де Age > 28)
print(f"Фільтрація .loc[df['Age'] > 28]:\n{df_from_dict.loc[df_from_dict['Age'] > 28]}\n")
```

```python
# 1. Імпортуємо бібліотеки
import matplotlib.pyplot as plt
# (NumPy та Pandas вже мають бути імпортовані)

# --- 1. Генерація даних ---
# Створимо набір 'X'
X = np.linspace(0, 20, 50) # 50 точок від 0 до 20

# Створимо ідеальну лінійну функцію Y_hat = m*X + b 
# Нехай m=2.5, b=5
Y_hat = 2.5 * X + 5

# Створимо "виміряні" дані (Y) з урахуванням помилок вимірювань 
# Додамо "шум" (випадкові помилки)
noise = np.random.normal(0, 3.0, X.shape) # Нормальний розподіл з середнім 0 та ст. відхиленням 3.0
Y = Y_hat + noise

print("Дані X, Y та Y_hat згенеровано.")


# --- 2. Побудова графіка --- [cite: 156]
plt.figure(figsize=(10, 6)) # Задаємо розмір малюнка

# Малюємо згенеровані дані (з помилками) як точки
plt.scatter(X, Y, label='Згенеровані дані (Y)', color='blue', alpha=0.7)

# Малюємо ідеальну функцію як лінію [cite: 93]
plt.plot(X, Y_hat, label='Лінійна функція (Y_hat = 2.5*X + 5)', color='red', linewidth=2)

# Додаємо оформлення [cite: 94-98]
plt.title('Графік лінійної функції та згенерованих даних')
plt.xlabel('Вісь X')
plt.ylabel('Вісь Y')
plt.legend()
plt.grid(True)
plt.show() # Показати графік


# --- 3. Обчислення метрик MAE та MSE ---
# Y - це наші "фактичні" (виміряні) значення (y_i) 
# Y_hat - це наші "передбачені" (ідеальні) значення (y_hat_i) 

# MAE (Mean Absolute Error) 
MAE = np.mean(np.abs(Y - Y_hat))

# MSE (Mean Squared Error) 
MSE = np.mean((Y - Y_hat)**2)

print(f"Обчислені метрики похибок:")
print(f"Mean Absolute Error (MAE): {MAE:.4f}")
print(f"Mean Squared Error (MSE): {MSE:.4f}")


# --- 4. Запис результатів у CSV --- [cite: 8, 158]
# Формат: X, Y, Y_hat, MAE, MSE

# Створюємо DataFrame для збереження
results_df = pd.DataFrame({
    'X': X,
    'Y': Y,
    'Y_hat': Y_hat
})

# Додаємо MAE та MSE (вони будуть однакові для кожного рядка)
results_df['MAE'] = MAE
results_df['MSE'] = MSE

# Зберігаємо у файл
csv_filename = 'lab2_results.csv'
results_df.to_csv(csv_filename, index=False)

print(f"Результати успішно збережено у файл: {csv_filename}")
print(results_df.head())
```
